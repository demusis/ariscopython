** Crie uma funÃ§Ã£o que conte quantas vezes a palavra 'gato' um texto possui. **

```python
def conteGatos(texto):
    contador = 0
    for palavra in texto.lower().split():
        if palavra == 'gato': contador = contador + 1
    return(contador)        
```

```python
conteGatos('Este gato e mais bravo que o seu gato')
```
