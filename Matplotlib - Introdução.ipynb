# Exemplo bÃ¡sico

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

## Comandos bÃ¡sicos

```python
plt.plot(x, y, 'r') # 'r' is the color red
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('TÃ­tulo')
plt.show()
```

## Criando vÃ¡rios grÃ¡ficos no mesmo espaÃ§o

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') # Veremos as opÃ§Ãµes de cor depois
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

___
# OrientaÃ§Ã£o a objeto

```python
# Cria uma figura vazia
fig = plt.figure()

# Adiciona eixos na figura
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # esquerda, inferior, largura, height (0-1)

# Eixos
axes.plot(x, y, 'b')
axes.set_xlabel('Seta o rÃ³tulo de X') # Observe o uso de set_ para iniciar os mÃ©todos
axes.set_ylabel('Seta o rÃ³tulo de Y')
axes.set_title('Seta o TÃ­tulo')
```

O cÃ³digo Ã© um pouco mais complicado, mas a vantagem Ã© que agora temos o controle total de onde os eixos da parcela sÃ£o colocados, e podemos facilmente adicionar mais de um eixo Ã  figura:

```python
# Cria um desenho em branco
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Eixos principais
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # Eixos internos

# Figura maior
axes1.plot(x, y, 'b')
axes1.set_xlabel('X1')
axes1.set_ylabel('Y1')
axes1.set_title('TÃ­tulo 1')

# Figura menor
axes2.plot(y, x, 'r')
axes2.set_xlabel('X2')
axes2.set_ylabel('Y2')
axes2.set_title('TÃ­tulo 2');
```

## subplots()

O objeto plt.subplots() Ã© um gerenciador de eixos mais automatizado.

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
# EspaÃ§o vazido de 1 x 2
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# Os eixos estÃ£o em arrays
axes
```

No qual podemos iterar:

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('TÃ­tulo')

# Mostra a figura    
fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('TÃ­tulo')

fig    
plt.tight_layout() # Evita sobreposiÃ§Ã£o
```

### Legendas

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

```python
# Alterando a posiÃ§Ã£o da legenda.....

ax.legend(loc=1) # superior direito
ax.legend(loc=2) # superior esquerdo
ax.legend(loc=3) # inferior esquerdo
ax.legend(loc=4) # inferior direito

ax.legend(loc=0) # Deixe o matplotlib decidir a posiÃ§Ã£o Ã³tima
fig
```

## Cores, espessuras, tipos

```python
# Ao estilo MATLAB..... 
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # Linha azul com pontos
ax.plot(x, x**3, 'g--') # Linha verde tracejada
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # Meio transparente
ax.plot(x, x+2, color="#8B008B")        # CÃ³digo hexadecimal RGB
ax.plot(x, x+3, color="#FF8C00")        # CÃ³digo hexadecimal RGB
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# â-â, âââ, â-.â, â:â, âstepsâ
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# Customizado
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # Formato: comprimento da linha, comprimento do espaÃ§o, ...

# Simbolos: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# Tamanho e cor
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
       markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

### Escala

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("PadrÃ£o")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("Justo")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("Customizado");
```
